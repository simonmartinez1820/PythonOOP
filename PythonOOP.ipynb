```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import output




class Notas:
    def __init__(self):
        self.ls = []
        self.df = None

    def randomList(self):
        ls = []

        for i in range(12):
            ls.append(round(random.uniform(0.0, 5.0), 1))


        self.df = pd.DataFrame(np.array(ls).reshape((3, 4)))
        self.df.columns = ['Pediatría', 'Internista', 'Neumología', 'Oftalmología']
        df = self.df
        turnos = ['Turno 1', 'Turno 2', 'Turno 3']
        df['Turnos'] = turnos

        df.set_index('Turnos', inplace=True)
        return df

    def printList(self):
        print(self.df)

    def notasMax(self):
        nota = []
        turnos = []
        especializacion = []
        for j in range(self.df.shape[0]):
            for i in range(self.df.shape[1]):
                if self.df.iloc[j, i] == self.df.max().max():
                    nota.append(self.df.iloc[j, i])
                    turnos.append(f'Turno {j + 1}')
                    especializacion.append(self.df.columns[i])

        df = pd.DataFrame(np.array([nota, turnos]).reshape((2, len(nota))))
        df.columns = especializacion
        df['Datos'] = ['Nota','Turno']
        df.set_index('Datos', inplace = True)

        return df

    def notasSobre4(self):
        nota = []
        turnos = []
        especializaciones = []
        for j in range(self.df.shape[0]):
            for i in range(self.df.shape[1]):
                if self.df.iloc[j, i] >= 4:
                    nota.append(self.df.iloc[j, i])
                    turnos.append(f'Turno {j + 1}')
                    especializaciones.append(self.df.columns[i])

        df = pd.DataFrame(np.array([nota, turnos]).reshape((2, len(nota))))
        df.columns = especializaciones
        df['Datos'] = ['Nota','Turno']
        df.set_index('Datos', inplace = True)
        return df

    def mean(self):
        print(round(self.df.values.mean(), 2))

    def specializationMean(self):
        print(dict(round(df.mean(), 2)))

    def promedioEspecializacion(self):
        ls = []
        for i in self.df.mean():
            ls.append(round(i, 2))
        df = pd.DataFrame(np.array(ls).reshape((1, 4)))
        df.columns = ['Pediatría', 'Internista', 'Neumología', 'Oftalmología']
        df['Dato'] = ['Promedio']
        df.set_index('Dato', inplace = True)
        return df

    def graficarPromedios(self):
        data = self.promedioEspecializacion()

        x=data.columns
        y=data.values[0]

        fig, ax = plt.subplots()
        bars = ax.bar(x, y)
        ax.set_xlabel("Especializaciones")
        ax.set_ylabel("Promedio")
        ax.set_title("Promedios por Especializacion")

        bars[0].set_facecolor('red')
        bars[1].set_facecolor('green')
        bars[2].set_facecolor('blue')
        bars[3].set_facecolor('orange')

        plt.show()

    def ordenarNotas(self):
        nota = []
        for j in range(self.df.shape[0]):
            for i in range(self.df.shape[1]):
                nota.append(self.df.iloc[j, i])

        for i in range(1, len(nota)):
            key = nota[i]
            j = i - 1
            while nota[j] > key and j >= 0:
                nota[j + 1] = nota[j]
                j -= 1
            nota[j + 1] = key
        return (nota)


print('Bienvenido al Sistema de Notas, posibles acciones:')
nota = Notas()
nota.randomList()
acciones = ['crear nueva lista', 'ordenar lista', 'notas maximas', 'notas sobre 4', 'promedio por especializacion',
            'graficar promedios por especializacion', 'imprimir lista','Promedio General','Para salir escribir "exit"']
ans = '0'
while ans != 'exit':
    for i in range(len(acciones)):
        print(f'{i + 1}. {acciones[i]}')
    print()

    if (int(ans) == 1):
        print(nota.randomList())
    elif (int(ans) == 2):
        print(nota.ordenarNotas())
    elif (int(ans) == 3):
        print(nota.notasMax())
    elif (int(ans) == 4):
        print(nota.notasSobre4())
    elif (int(ans) == 5):
        print(nota.promedioEspecializacion())
    elif (int(ans) == 6):
        nota.graficarPromedios()
    elif (int(ans) == 7):
        nota.printList()
    elif(int(ans)==8):
      print(nota.mean())

    print()

    ans = input()
    output.clear()

```
